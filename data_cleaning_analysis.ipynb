Grouping by multiple groups gives...

```python
df_gptest = df[['drive-wheels', 'body-style','price']]
```

```python
df_gptest = df_gptest.groupby(['drive-wheels','body-style'],as_index=False).mean()
df_gptest
```

The pivot method can be used to convert the previous table into a pivot table

```python
grouped_pivot = df_gptest.pivot(index='drive-wheels',columns='body-style')
grouped_pivot
```

```python
grouped_pivot = grouped_pivot.fillna(0)
grouped_pivot
```

```python
groupby_2 = df[['body-style','price']]
groupby_2 = groupby_2.groupby(['body-style'], as_index=False).mean()
groupby_2
```

Creating a heatmap of the pivot table

```python
plt.pcolor(grouped_pivot, cmap='RdBu')
plt.colorbar()
plt.show()
```

```python
from scipy import stats
```

```python
df[['wheel-base','price']].corr()
```

```python
pearson_coef, p_value = stats.pearsonr(df['wheel-base'],df['price'])
print("The pearson coef is {} and its p-value is {}".format(pearson_coef,p_value))
```

```python
stats.pearsonr(df['price'],df['horsepower'])
```

Large pearson coefficient, and p-value very close to 0. This implies a positive linear correlation between price and horsepower and horsepower is statistically significant in determining the price.

```python
pearson_coef, p_value = stats.pearsonr(df['bore'], df['price'])
print("The pearson coef is {} and its p-value is {}.".format(pearson_coef, p_value))
```

While statistically significant, there is not a solid relationship between boresize and price
